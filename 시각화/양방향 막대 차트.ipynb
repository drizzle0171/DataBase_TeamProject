```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'NanumGothic'
```

```python
#인천 vs 서울
#데이터 순서: 의료, 주거, 교통, 교육, 상권
incheon = np.array([2227, 696, 1768, 4039, 20640])
seoul = np.array([5841, 1943, 2155, 6716, 23604])
X = np.arange(5)

plt.barh(X, incheon)
plt.barh(X, -seoul)
ticklabel = ['Medical', 'Housing', 'Traffic', 'Education', 'Commercial district']
plt.yticks(X, ticklabel)
plt.ylabel('Infrastructure')
plt.title('Seoul vs Incheon')
plt.show()
```

```python
#인천 vs 서울
#데이터 순서: 고용율, 매매지수
incheon_rate = np.array([54.93, 84.53])
seoul_rate = np.array([55.93, 86.78])

Y = np.arange(2)
plt.barh(Y, incheon_rate)
plt.barh(Y, -seoul_rate)
ticklabel2 = ['Employment rate', 'Sales index']
plt.yticks(Y, ticklabel2)
plt.ylabel('Rate')
plt.title('Seoul vs Incheon')
plt.show()
```
