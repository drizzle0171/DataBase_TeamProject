```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#인천 vs 서울
#데이터 순서: 의료, 주거, 교통, 교육, 상권, 고용율, 매매지수
incheon = np.array([2227, 696, 1768, 4039, 20640, 54.93, 84.53])
seoul = np.array([5841, 1943, 2155, 6716, 23604, 55.93, 86.78 ])
X = np.arange(7)

plt.barh(X, incheon)
plt.barh(X, -seoul)
plt.show()
```
